```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data setup
models = ['Vosk', 'Whisper Small', 'Whisper Fine Tune']
wer = [39.48, 22.93, 12.62]  # Word Error Rate
cer = [22.31, 17.18, 2.67]   # Character Error Rate

# Ensure plots display inline in Google Colab
%matplotlib inline
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for WER
ax1.bar(models, wer, color='lightblue', label='WER', width=0.4)
ax1.set_ylabel('Word Error Rate (%)')
ax1.set_title('WER and CER Comparison')

# Line chart for CER
ax2 = ax1.twinx()
ax2.plot(models, cer, color='darkgreen', marker='o', label='CER')
ax2.set_ylabel('Character Error Rate (%)')

# Final touches
fig.legend(loc="upper left", bbox_to_anchor=(0.1,0.9))
plt.show()
```

```python
# Prepare data for heatmap

import pandas as pd
heatmap_data = np.array([wer, cer]).T
heatmap_df = pd.DataFrame(heatmap_data, index=models, columns=['WER', 'CER'])

plt.figure(figsize=(8, 3))
sns.heatmap(heatmap_df, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of WER and CER')
plt.show()
```

```python
import plotly.graph_objects as go

# Data
models = ['Vosk', 'Whisper Small', 'Whisper Fine Tune']
wer = [39.48, 22.93, 12.62]
cer = [22.31, 17.18, 2.67]

# Create traces
fig = go.Figure()
fig.add_trace(go.Bar(x=models, y=wer, name='WER'))
fig.add_trace(go.Bar(x=models, y=cer, name='CER'))

# Update the layout
fig.update_layout(title_text='Model Performance', xaxis_title='Model', yaxis_title='Error Rate (%)', barmode='group')

fig.show()
```
